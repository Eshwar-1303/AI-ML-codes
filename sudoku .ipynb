```python
board = [[0,0,0,0,8,4,0,0,3],
        [1,3,0,0,5,0,0,9,4],
        [9,0,5,0,1,3,0,7,0],
        [4,0,0,0,9,6,0,0,0],
        [0,0,3,7,0,0,0,0,1],
        [7,5,0,0,3,0,4,6,9],
        [5,0,4,0,7,8,0,2,0],
        [3,6,0,0,2,9,0,5,0],
        [0,2,0,0,0,0,3,0,7]]


def solve(bo):
    find = find_empty(bo)
    if not find:
        return True
    else:
        row, col = find

    for i in range(1,10):
        if valid(bo, i, (row, col)):
            bo[row][col] = i

            if solve(bo):
                return True

            bo[row][col] = 0

    return False


def valid(bo, num, pos):
    # Check row
    for i in range(len(bo[0])):
        if bo[pos[0]][i] == num and pos[1] != i:
            return False

    # Check column
    for i in range(len(bo)):
        if bo[i][pos[1]] == num and pos[0] != i:
            return False

    # Check box
    box_x = pos[1] // 3
    box_y = pos[0] // 3

    for i in range(box_y*3, box_y*3 + 3):
        for j in range(box_x * 3, box_x*3 + 3):
            if bo[i][j] == num and (i,j) != pos:
                return False

    return True


def find_empty(bo):
    for i in range(len(bo)):
        for j in range(len(bo[0])):
            if bo[i][j] == 0:
                return (i, j)  # row, col

    return None


solve(board)
print(board)
```
